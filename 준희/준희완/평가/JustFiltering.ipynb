```python
import numpy as np 
import pandas as pd
import json
```

```python
#웹에서 가져온 데이터 필터값
hyo = '1'
caff ='X'
hyang = '2'
```

```python
meta = pd.read_csv('teaable.csv')

meta
```

```python
df = pd.DataFrame(meta)
df
```

```python
df.dtypes
```

```python
df.insert(2,'splithyo',df.효능번호.str.split(','))
df.insert(5,'splithyang',df.향기준.str.split(','))
df
```

```python
df.dtypes
```

```python
def hyo_check(nations):
    n_list = []
    
    for nation in nations:
        if nation == hyo:
            n_list.append(True)
        else:
            n_list.append(False)
            
    if np.sum(n_list)>=1:
        return True
    
    else :
        return False
```

```python
def hyang_check(nations):
    n1_list = []
    
    for nation in nations:
        if nation == hyang:
            n1_list.append(True)
        else:
            n1_list.append(False)
            
    if np.sum(n1_list)>=1:
        return True
    
    else :
        return False
```

```python
df.splithyo[:10].apply(hyo_check)
```

```python
df.splithyang.apply(hyang_check)
```

```python
df.splithyo.apply(hyo_check).sum()
#검색필터 개수(효능)
```

```python
(df.효능번호 =='1').sum()
#검색필터 개수(효능 중에 "," 값 제외)
```

```python
df.splithyang.apply(hyang_check).sum()
#검색필터 개수(향)
```

```python
#필터링 조건

cond = (df.splithyo.apply(hyo_check )) 
cond1 = (df.splithyang.apply(hyang_check ))
cond2 = df.카페인==caff


Kr_df = df[cond &cond1 &cond2]



print(Kr_df.shape)
#필터링출력값
Kr_df



```
